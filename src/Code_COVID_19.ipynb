```python
def normality_check(data):
    pos= []
    th_Q = []
    data = np.sort(np.array(data))
    pos = [(i-0.5)/len(data) for i in range(1,len(data)+1)]
    th_Q = [norm.ppf(i,np.mean(data),np.std(data,ddof = 1)) for i in pos]
    #print(th_Q)
    plt.plot(data,th_Q,'ro',data,data)
    plt.show()
    
    
normality_check(India_df['Cured'][220:250])
normality_check(Karnataka_df['Cured'][200:])
```

```python
import random
import statistics
import math

mean_cases = df['Daily_cases'].tail(100).mean()
print("Mu value")
print(mean_cases)

std_cases = df['Daily_cases'].tail(100).std()

sample = df['Daily_cases'].tail(100)
sampled_output = random.sample(list(sample), 45)

Xbar = statistics.mean(sampled_output)
s = statistics.stdev(sampled_output)
print("Xbar")
print(Xbar)
print("Sample deviation")
print(s)
z = 1.645

CP = (z*(s/ math.sqrt(100))) + mean_cases
print("Critical Point")
print(CP)

if Xbar>=CP:
     print("reject null hypothesis")
else:
    print("Doesnt have enough evidence to reject null hypothesis.Both are plausible.")
```

Chi Squared test - Test of Goodness of fit (Single Categorical Variable)

H0 : All States have had an equal contribution of confirmed cases On 27th October 
H1 : All States do not contribute equally to confirmed cases on 27th October

```python
import scipy.stats as stats
dataset =pd.read_csv(r'C:\Users\91944\Downloads\clean_data_covid.csv')
#perform Chi-Square Goodness of Fit Test

data = dataset['Confirmed'].tail(35)
#data

observed = data

expected = [data.sum()/35]*35

stat,p = stats.chisquare(f_obs=observed, f_exp=expected)

if(p< 0.05):
    print("Reject Null Hypothesis")
else:
    print("Both H0 and H1 are Plausible")
```
